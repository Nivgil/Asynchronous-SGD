```python
import numpy as np
import pandas as pd
from bokeh.io import output_notebook
from bokeh.plotting import figure, show, output_file
from Outputs.jupyter_graphs import create_graphs, compare_graphs
from IPython.display import Markdown, display
from IPython.core.interactiveshell import InteractiveShell
import warnings
warnings.filterwarnings('ignore')
sim_num = [201, 202, 103,209,210]
plots, scores = compare_graphs(sim_nums=sim_num, linear=True)
output_notebook()
for sim in zip(sim_num,scores):
    display(Markdown('<h6> #Simulation {:1d} - Min Error {:.2f}% , Mean Error {:.2f}% , Val Error {:.2f}%<h6>'.format(sim[0],sim[1][0],sim[1][1],sim[1][2])))
```

The simulation examines the trade-off between mini batch size and number of workers in a synchrnous / asynchronous setting.

The number of epochs is the same for all experiments. 

(BLUE) Simulation 201 - Synchronous, 1 worker, 256 batch size + learning rate adjustment 

(ORANGE) Simulation 202 - Synchronous, 16 workers, 16 batch size + learning rate adjustment  (technically it is 1 worker with ghost batch norm of 16)

(GREEN) Simulation 103 - Asynchronous, 16 workers, 16 batch size 

(RED) Simulation 209 - Synchronous, 16 workers, 256 batch size + learning rate adjustment + gbn (technically it is 1 worker with batch size of 4096)

(PURPLE) Simulation 210 - Asynchronous, 16 workers, 256 batch size + learning rate adjustment + gbn + momentum 0
